### Word Enbedding using Keras

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am good developer',
    'understand the meaning of words',
    'the videos is good'
]
```

```python
sent
```

```python
# define vocabulary size
voc_size=10000
```

```python
## OHE Representation
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
#Word Embedding Representation
from tensorflow.keras.layers import Embedding

from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
'''
2️⃣ Padding Sequences
What it does
This function makes all sequences the same length.
Neural networks require inputs with equal size.
Example sequences:
[1, 4, 6]
[2, 3]
[5, 7, 8, 9]
After padding:
[1, 4, 6, 0]
[2, 3, 0, 0]
[5, 7, 8, 9
'''
from tensorflow.keras.models import Sequential
'''
3️⃣ Sequential Model
from tensorflow.keras.models import Sequential
What it does

This creates a Sequential Neural Network model.

Sequential means layers are added one after another.

Structure:

Input → Layer1 → Layer2 → Layer3 → Output
'''
```

```python
# applying padding in ohe words to convert it in equal words 
sent_lenght=8
emd_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_lenght) # i am applying pre means i want to de pre padding you can also apply post.
```

```python
emd_docs
```

```python
## feature repreestation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_lenght))
model.compile('adam','mse')
'''In TensorFlow / Keras, compile() means preparing the model for training.
It tells the model:

Which optimizer to use → how weights will be updated

Which loss function to use → how error will be calculated

Which metrics to track (optional)'''
```

```python
model.predict(emd_docs)
```
